## Which company is the most common video game publisher?

```python
#scott
#df['Publisher'].groupby(df['Publisher']).count().sort_values(ascending=False).head(1)
df.Publisher.groupby(df.Publisher).count().sort_values(ascending=False).head(2) # can use dot notation unless column has space
# get column publisher
# group by and count collapses
```

```python
df[df.Publisher == 'Electronic Arts'].head(5) #return all games where electronic arts is the publisher
```

```python
#publisher = pd.DataFrame(df['Publisher']).groupby(df.Publisher).count()
publisher = pd.DataFrame(df.Publisher).groupby(df.Publisher).count()
publisher = publisher.sort_values(['Publisher'], ascending=False)
publisher.head()
```

## What is the most common platform?

```python
#scott
#df.Platform.groupby(df.Platform).count().sort_values(ascending=False).head()
platform = pd.DataFrame(df['Platform']).groupby(df.Platform).count()
platform = platform.sort_values(['Platform'], ascending=False)
platform.head()
```

## What is the most common genre?

```python
#scott
# df.Genre.groupby(df.Genre).count().sort_values(ascending=False).head()
#df.info() lets us know data type of values
#df.describe() lets you know mean, std, other info of each column
genre = pd.DataFrame(df['Genre']).groupby(df.Genre).count()
genre = genre.sort_values(['Genre'], ascending=False)
genre.head()
```
